```python
import sys
sys.path.append('../../..')

from batchflow.research import Experiment, Research, Option, Domain, EC, O, EP

def f(x, y):
    return x + y

experiment = (Experiment()
    .add_callable('sum', f, x=EC('x'), y=EC('y'))
    .save(O('sum'), 'sum')
)

domain = Domain(x=[1, 2], y=[2, 3, 4])
research = Research(name='new_research', experiment=experiment, domain=domain)
research.remove()
research.run(n_iters=10)

# from batchflow.research import Domain, Research, RC

# domain = Domain({'x': [1, 2], 'y': [2, 3, 4]})

# def f(config):
#     config = config.config()
#     x = config['x']
#     y = config['y']
#     return x + y

# research = (Research()
#             .init_domain(domain)
#             .add_callable(f, config=RC(), name='f', returns='a', dump=5))

# ! rm -r old_research
# research.run(name='old_research', n_iters=10)
```
